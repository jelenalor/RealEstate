```python
import numpy as np 
import pandas as pd
import os
import psycopg2

import importlib
from Secrets import secrets
# importlib.reload(secrets)

import matplotlib.pyplot as plt
```

### Pull Data from a custom made AWS Database of public data
Define an engine to query AWS DB and set up the execution

```python
# Define Engine
engine = psycopg2.connect(
    database="postgres",
    user=secrets.user(),
    password=secrets.password(),
    host=secrets.host(),
    port='5432'
)

cur = engine.cursor()
```

```python
cur.execute("""

SELECT 
    lsoa, EXTRACT(YEAR FROM date), cat_group, sum(value)

from crime

group by 1, 2, 3;




""")
rows = cur.fetchall()
result = pd.DataFrame(rows, columns = [desc[0] for desc in cur.description])
result
```

```python
fig, ax = plt.subplots(figsize = (10, 4))

df = result.groupby("cat_group")["sum"].sum().reset_index()
width = 0.35
labels = df["cat_group"].to_list()
x = np.arange(len(labels))
y = df["sum"].to_list()

ax.bar(x, y, width)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of crime cases')
ax.set_title('Num of crime cases by category')
ax.set_xticks(x)
ax.set_xticklabels(labels)

plt.show()
```
